eq1 - mass conservation) ∂ρ/∂t+∇⋅(ρu)=0

eq2 - momentum conservation) ∂(ρu)/∂t+∇⋅(ρuu)=ρf+∇⋅σ

eq 3 - energy conservation) ρu*dx/dh=q′

For heat exchanger m˙*dx/dh=UP(Twall−Tfluid)

Assumptions:

Steady state - ∂/∂t = 0
1 dimensional - only along x axis (d/dy = d/dz = 0)
No body forces - pf = 0
Fully developed flow - friction factor can be used in eq2
Changes in velocity small relative to friction losses - (d/dx)(ρu^2)≪dxdp therefore (d/dx)*(pu^2) can be ignored

This gives (rearranged to = 0, all terms on LHS):

- eq1: a constant (ignored by PINN)

- eq2:  dp_dx + f * rho * u**2 / (2 * D) = 0

- eq3: m˙*dx/dh - U*P*(Twall−Tfluid) = 0
